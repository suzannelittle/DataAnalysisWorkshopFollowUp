So doc contains 10 strings, one for each page (or year) in the PDF document in reverse order since the first entry contains '1969'. Looks a bit messy but if we split the string into words and remove the header information (the first 19 words) then it's basically: Position, BoyName, NumBirths, GirlName, NumBirths.

````python
words = doc[0].split()
words[19:29]
````

David (2525) and Michelle (1453) were the top ranked names in 1969 while Glen (204) and Anna (215) were ranked at 51.

We can loop through the words in each string to extract the data, put into a dataframe and save it as csv. Recall the US baby names were structured as `[name, sex, births, year]` so let's copy that. We can get the year from the word at position 9.

````python
words[9]
````

We need to start at 19 to get the first position and process the next 5 words. We don't need position (this can be calculated from the number of births), we want to end up with (name,sex,births,year)

````python
data = []
for x in range(19,len(words),5):
    entry = words[x:x+5]
    data.append((entry[1], 'M', entry[2], words[9]))
    data.append((entry[3], 'F', entry[4], words[9]))
df = pd.DataFrame(data, columns=['name', 'sex', 'births', 'year'])
df.head(5)
````

Connecting all that together to loop over all documents and pages and get a big dataframe.

````python
import os
location = '/home/slittle/data/text/nswbabynames/'
documents = os.listdir(location)
dfs = []
for filename in documents:
    with open(os.path.join(location,filename)) as fh:
        doc = slate.PDF(fh)
    data = []
    for page in doc:
        words = page.split()
        start = words.index('1')  # 2010s has extra text which changes position
        # 1998 is missing the position before boys column (exc 1)
        if words[start-10] == '1998': 
            problem = words
            continue  # too hard to fix right now!
        for w in range(start,start+250,5):
            entry = words[w:w+5]
            try:
                data.append((entry[1], 'M', int(entry[2]), words[start-10]))
                data.append((entry[3], 'F', int(entry[4]), words[start-10]))
            except Exception as e:
                print filename, start, words[start-10], w, entry
                raise e
    dfs.append(pd.DataFrame(data, columns=['name', 'sex', 'births', 'year']))
    print dfs[-1].head(5)
````
